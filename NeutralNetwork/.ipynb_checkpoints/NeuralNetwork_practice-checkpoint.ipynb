```python
import torch
class MY_MNIST():
    def __init__(self, X_train,label):
        self.data=X_train
        self.targets = label
    def __getitem__(self, index):
        img, target = self.data[index], int(self.targets[index])
        img = torch.tensor(img,dtype=torch.float)
        sample = {'img' : img, 'target': target}
        return sample
    def __len__(self):
        return len(self.data)
```

```python
train_data = MY_MNIST(X_train,Y_train)
print(train_data[0]['img'].shape)
print(train_data[3]['target'])
```

```python
from torch.utils.data import random_split
train_size = int(0.8 * len(train_data))
val_size = len(train_data) - train_size
train_dataset, val_dataset = random_split(train_data,lengths=[train_size,val_size],generator=torch.Generator().manual_seed(42))
```

```python
from torch.utils.data import DataLoader
train_dataloader = DataLoader(train_dataset,batch_size=30,shuffle=True)
```

```python
import torch.nn as nn
import torch 
class DigitModel(nn.Module):
    def __init__(self,in_channels=1,classes=10):
        super().__init__()
        self.feature = nn.Sequential(
            nn.Conv2d(in_channels,out_channels=10,kernel_size=(5,5)),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2,2)),
            nn.Conv2d(10,20,kernel_size=(5,5)),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2,2)),
        )
        self.classifier = nn.Sequential(
            nn.Linear(320,50),
            nn.ReLU(),
            nn.Dropout(0.25),
            nn.Linear(50,classes)
        )
    def forward(self,x):
        x = self.feature(x)
        x = x.view(-1,320)
        x = self.classifier(x)
        return x
                     

            

                 
```

```python
model = DigitModel()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(),lr=1e-3,momentum=0.9)
```

```python
def train(epoch):
    running_loss = 0.0
    running_total = 0
    running_correct = 0
    for batch_idx, data in enumerate(train_dataloader):
        inputs, target = data['img'],data['target']
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, target)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        predicted = torch.argmax(torch.softmax(outputs,dim=1),dim=1)
        running_total += inputs.shape[0]
        running_correct += (predicted == target).sum().item()
        if batch_idx % 500 == 499:
            print(f"""[{epoch + 1}, {batch_idx +1:5d}]: 
                loss: {running_loss/running_total:.3f},
                acc: {100 * running_correct / running_total:.2f} %""")
    return running_loss/running_total
            
        
        
        
```

```python
test_loader = DataLoader(val_dataset,10,shuffle=True)
```

```python
def test():
    correct = 0
    total = 0
    with torch.no_grad():
        for data in test_loader:
            inputs, labels = data['img'], data['target']
            prec = model(inputs)
            predicted = torch.argmax(torch.softmax(prec,dim=1),dim=1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    print(f'accuracy on test set: {100 * correct / total} %')
    return correct/total    
        
```

```python
loss = []
accuracy = []
for i in range(10):
    loss.append(train(i))
    accuracy.append(test())
    
```

```python
plt.plot(loss, color='b', label="traing loss")
plt.xlabel("Number of Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
plt.plot(accuracy, color='b',label="accuracy")
plt.xlabel("Number of Epochs")
plt.ylabel("acc")
plt.legend()
plt.show()
```

```python
from PIL import Image, ImageOps

def resize_image(src_image, size=(128,128),bg_color="white"):
    src_image.thumbnail(size, Image.ANTIALIAS)
    new_image = Image.new("RGB", size, bg_color)
    new_image.paste(src_image, (int((size[0] - src_image.size[0]) / 2), int((size[1] - src_image.size[1]) / 2)))
    return new_image
```

```python
training_folder_name = './archive/data/natural_images'

# 存储处理后的图片
train_folder = './archive/working/data/natural_images'
size = (128,128)
if not os.path.exists(train_folder):
    os.makedirs(train_folder)
print('转换图像...')
for root, folders, files in os.walk(training_folder_name):
    for sub_folder in folders:
        print('processing folder ' + sub_folder)
        # 创建子文件夹
        saveFolder = os.path.join(train_folder,sub_folder)
        if not os.path.exists(saveFolder):
            os.makedirs(saveFolder)
        # 遍历改子文件夹下的所有图片
        file_names = os.listdir(os.path.join(root,sub_folder))
        for file_name in file_names:
            #图片路径
            file_path = os.path.join(root,sub_folder, file_name)
            #打开图片
            image = Image.open(file_path)
            # 创建缩放版本并保存
            resized_image = resize_image(image, size)
            saveAs = os.path.join(saveFolder, file_name)
            resized_image.save(saveAs)

print('Done.')
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
def load_dataset(data_path):
    
    # 随机修改图像，防止过拟合
    transformation = transforms.Compose([
        # 随机水平翻转
        transforms.RandomHorizontalFlip(0.5),
        # 随机垂直反转
        transforms.RandomVerticalFlip(0.3),
        # 转变为tensor类型
        transforms.ToTensor(),
        # 标准化
        transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
    ])

    # 加载所有图像，并进行修改
    #将同一文件夹下的图片视为同一类，含有图片数据以及标签值
    full_dataset = torchvision.datasets.ImageFolder(
        root=data_path,
        transform=transformation
    )
    
    
    # 划分数据
    train_size = int(0.7 * len(full_dataset))
    test_size = len(full_dataset) - train_size
    train_dataset, test_dataset = torch.utils.data.random_split(full_dataset, [train_size, test_size])
    
    train_loader = torch.utils.data.DataLoader(
        train_dataset,
        batch_size=50,
        num_workers=0,
        shuffle=False
    )
    
    test_loader = torch.utils.data.DataLoader(
        test_dataset,
        batch_size=50,
        num_workers=0,
        shuffle=False
    )
        
    return train_loader, test_loader
    
```

```python
train_folder = './archive/working/data/natural_images'

train_loader, test_loader = load_dataset(train_folder)
batch_size = train_loader.batch_size
```

```python
class Net(nn.Module):
    def __init__(self,num_classes=3):
        super().__init__()
        
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=12, kernel_size=3, stride=1, padding=1)

        self.conv2 = nn.Conv2d(in_channels=12, out_channels=24, kernel_size=3, stride=1,padding=1)

        self.pool = nn.MaxPool2d(kernel_size=2)

        self.drop = nn.Dropout2d(p=0.2)

        self.fc = nn.Linear(in_features=32*32*24, out_features=num_classes)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = self.drop(x)
        x = x.view(-1, 32 * 32 * 24)
        x = self.fc(x)
        return x
device = "CPU"
if (torch.cuda.is_available()):
    device = "cuda"

model = Net(num_classes=len(classes)).to(device)

print(model)

          
```

```python
def train(model, device, train_loader, optimizer, epoch):
    model.train()
    train_loss = 0
    print("Epoch:", epoch)
    for batch_idx, (data, target) in enumerate(train_loader):

        data, target = data.to(device), target.to(device)
        
        # Reset the optimizer
        optimizer.zero_grad()

        #计算模型输出
        output = model(data)
        
        # 损失
        loss = loss_criteria(output, target)

        # 记录损失
        train_loss += loss.item()
        
        # 反向传播
        loss.backward()
        optimizer.step()
        if batch_idx % 10==9:
            print('\tTraining batch {} Loss: {:.6f}'.format(batch_idx + 1, loss.item()))
            
    # 每个epoch的平均损失
    avg_loss = train_loss / (batch_idx+1)
    print('Training set: Average loss: {:.6f}'.format(avg_loss))
    return avg_loss
```

```python
def test(model, device, test_loader):
    #测试时不需要反向传播与drop
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        batch_count = 0
        for data, target in test_loader:
            batch_count += 1
            data, target = data.to(device), target.to(device)
            
            # 每个批次的输出
            output = model(data)
            
            # 每个批次的损失
            test_loss += loss_criteria(output, target).item()
            
            # 该批次的预测正确个数
            predicted = torch.argmax(torch.softmax(output,dim=1), 1)
            correct += torch.sum(target==predicted).item()

    # 该epoch的平均损失与正确率
    avg_loss = test_loss / batch_count
    print('Validation set: Average loss: {:.6f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        avg_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
    
    return avg_loss
```

```python
#指定优化器
optimizer = optim.Adam(model.parameters(), lr=0.01)

# 指定损失函数
loss_criteria = nn.CrossEntropyLoss()

epoch_nums = []
training_loss = []
validation_loss = []

# Train over 10 epochs (We restrict to 10 for time issues)
epochs = 10
print('Training on', device)
for epoch in range(1, epochs + 1):
        train_loss = train(model, device, train_loader, optimizer, epoch)
        test_loss = test(model, device, test_loader)
        epoch_nums.append(epoch)
        training_loss.append(train_loss)
        validation_loss.append(test_loss)
```

```python
plt.figure(figsize=(8,8))
plt.plot(epoch_nums, training_loss)
plt.plot(epoch_nums, validation_loss)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training', 'validation'], loc='upper right')
plt.show()
```

```python
#Defining Labels and Predictions
truelabels = []
predictions = []
model.eval()
print("Getting predictions from test set...")
for data, target in test_loader:
    data = data.to(device)
    for label in target.data.numpy():
        truelabels.append(label)
    output = torch.softmax(model(data),dim=1)
    for prediction in torch.argmax(output,dim=1):
        predictions.append(prediction.cpu().numpy()) 

# Plot the confusion matrix, 将每一样本预测为各个类的情况
cm = confusion_matrix(truelabels, predictions)
tick_marks = np.arange(len(classes))

df_cm = pd.DataFrame(cm, index = classes, columns = classes)
plt.figure(figsize = (7,7))
sns.heatmap(df_cm, annot=True, cmap=plt.cm.Blues, fmt='g')
plt.xlabel("Predicted Shape", fontsize = 20)
plt.ylabel("True Shape", fontsize = 20)
plt.show()
```
