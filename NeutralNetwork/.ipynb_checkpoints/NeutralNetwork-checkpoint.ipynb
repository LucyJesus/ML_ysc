卷积层进行的处理就是卷积运算。卷积运算相当于图像处理中的“滤波器运算”

对于输入数据，卷积运算以一定间隔滑动滤波器的窗口并应用。将各个位置上滤波器的元素和输入的对应元素相乘，然后再求和（有时将这个计算称为乘积累加运算）。然后，将这个结果保存到输出的对应位置。将这个过程在所有位置都进行一遍，就可以得到卷积运算的输出。


![avatar](../imgs/2.jpg)

![title](../imgs/3.jpg)

在进行卷积层的处理之前，有时要向输入数据的周围填入固定的数据（比如0等），这称为填充，是卷积运算中经常会用到的处理。如图

![avatar](../imgs/4.jpg)

使用填充主要是为了调整输出的大小。比如，对大小为(4, 4)的输入数据应用(3, 3)的滤波器时，输出大小变为(2, 2)，相当于输出大小比输入大小缩小了2个元素。如果每次进行卷积运算都会缩小空间，那么在某个时刻输出大小就有可能变为1，导致无法再应用卷积运算。为了避免出现这样的情况，就要使用填充。

应用滤波器的位置间隔称为步幅（stride）。之前的例子中步幅都是1，如果将步幅设为2，应用滤波器的窗口的间隔变为2个元素。

![avatar](../imgs/5.jpg)

综上，增大步幅后，输出大小会变小。而增大填充后，输出大小会变大。

假设输入大小为(H, W)，滤波器大小为(FH, FW)，输出大小为(OH, OW)，填充为P，步幅为S。此时，输出大小可通过下式进行计算。

![avatar](../imgs/6.jpg)

之前的卷积运算的例子都是以有高、长方向的2维形状为对象的。但是，图像是3维数据，除了高、长方向之外，还需要处理通道方向。

需要注意的是，在3维数据的卷积运算中，输入数据和滤波器的通道数要设为相同的值。
![avatar](./7.jpg)

通道方向上有多个特征图时，会按通道进行输入数据和滤波器的卷积运算，并将结果相加，从而得到输出

![avatar](../imgs/8.jpg)

神经网络的处理中进行了将输入数据打包的批处理。通过批处理，能够实现处理的高效化和学习时对mini-batch的对应。

我们希望卷积运算也同样对应批处理。为此，需要将在各层间传递的数据保存为4维数据。具体地讲，就是按(batch_num, channel, height, width)的顺序保存数据。

![avatar](../imgs/9.jpg)

池化是缩小高、长方向上的空间的运算。如图所示，进行将2 × 2的区域集约成1个元素的处理，缩小空间大小。

![avatar](../imgs/10.jpg)

图中是按步幅2进行2 × 2的Max池化时的处理顺序。“Max池化”是获取最大值的运算，“2 × 2”表示目标区域的大小。

除了Max池化之外，还有Average池化等。相对于Max池化是从目标区域中取出最大值，Average池化则是计算目标区域的平均值。在图像识别领域，主要使用Max池化。池化层有以下特征：没有要学习的参数池化层和卷积层不同，没有要学习的参数。池化只是从目标区域中取最大值（或者平均值），所以不存在要学习的参数。通道数不发生变化经过池化运算，输入数据和输出数据的通道数不会发生变化。计算是按通道独立进行的。

![avatar](../imgs/11.jpg)


卷积的作用就是为了提取某些指定的特征，而池化就是为了进一步抽取更高阶的特征。通过池化操作忽略一些细节信息，强行让CNN学到的更多我们想要的高阶信息。

激活函数（Activation Function）是一种添加到人工神经网络中的函数，旨在帮助网络学习数据中的复杂模式。在神经元中，输入的input经过一系列加权求和后作用于另一个函数，这个函数就是这里的激活函数。类似于人类大脑中基于神经元的模型，激活函数最终决定了是否传递信号以及要发射给下一个神经元的内容。

因为神经网络中每一层的输入输出都是一个线性求和的过程，下一层的输出只是承接了上一层输入函数的线性变换，所以如果没有激活函数，那么无论你构造的神经网络多么复杂，有多少层，最后的输出都是输入的线性组合，纯粹的线性组合并不能够解决更为复杂的问题。而引入激活函数之后，我们会发现常见的激活函数都是非线性的，因此也会给神经元引入非线性元素，使得神经网络可以逼近其他的任何非线性函数，这样可以使得神经网络应用到更多非线性模型中。

```python
X_train = X_train / 255.0
test = test / 255.0
print("x_train shape: ",X_train.shape)
print("test shape: ",test.shape)
```

```python
X_train = X_train.values.reshape(-1,1,28,28)
test = test.values.reshape(-1,1,28,28)
print("x_train shape: ",X_train.shape)
print("test shape: ",test.shape)
```

```python
import torch
class MY_MNIST():
    def __init__(self, X_train,label):
        self.data=X_train
        self.targets = label
    def __getitem__(self, index):
        img, target = self.data[index], int(self.targets[index])
        img = torch.tensor(img,dtype=torch.float)
        sample = {'img': img, 'target': target}
        return sample
    def __len__(self):
        return len(self.data)
```

```python
train_data = MY_MNIST(X_train,Y_train)
print(train_data[0]['img'].shape)
print(train_data[3]['target'])
```

```python
from torch.utils.data import random_split
train_size = int(0.8 * len(train_data))      # full_dataset是整个数据集,取0.8的比例作为训练集
val_size = len(train_data) - train_size
train_dataset,val_dataset = random_split(train_data,lengths=[train_size,val_size],generator=torch.Generator().manual_seed(42))
```

```python
from torch.utils.data import  DataLoader
#处理模型输入数据的一个工具类。组合了数据集（dataset） + 采样器(sampler)，
# 并在数据集上提供单线程或多线程(num_workers )的可迭代对象。
train_dataloader = DataLoader(train_dataset,batch_size=30,shuffle=True)
```

```python
#搭建卷积神经网络模型
import torch.nn as nn
import torch
class DigitModel(nn.Module):
    def __init__(self,in_channels=1,classes=10):
        super().__init__()
        self.feature = nn.Sequential(
            nn.Conv2d(in_channels,out_channels=10,kernel_size=(5,5)),#28x28->24x24
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2,2)),#24x24->12x12
            nn.Conv2d(10,20,kernel_size=(5,5)),#12x12->8x8
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2,2)),#8x8->4x4
            
        )
        self.classifier = nn.Sequential(
            nn.Linear(320,50),
            nn.ReLU(),
            nn.Dropout(0.25),#防止出现过拟合
            nn.Linear(50,classes)
        )
    def forward(self,x):
        x = self.feature(x)
        x = x.view(-1,320)
        #根据前面提取到的特征进行分类
        x = self.classifier(x)
        return x

```

```python
#实例化模型
model = DigitModel()
#损失函数
criterion = nn.CrossEntropyLoss()
#优化器
optimizer = torch.optim.SGD(model.parameters(),lr=1e-3,momentum=0.9)
```

```python
def train(epoch):
    running_loss = 0.0  # 这整个epoch的loss清零
    running_total = 0
    running_correct = 0
    for batch_idx, data in enumerate(train_dataloader):
        inputs, target = data['img'],data['target']
        optimizer.zero_grad()

        # forward + backward + update
        outputs = model(inputs)
        loss = criterion(outputs, target)

        loss.backward()
        optimizer.step()

        # 记录损失
        running_loss += loss.item()
        # 记录准确率
        predicted = torch.argmax(torch.softmax(outputs, dim=1),dim=1)#返回输出中最大的一项及其对应的索引值
        running_total += inputs.shape[0]
        running_correct += (predicted == target).sum().item()

        if batch_idx % 500 == 499:  # 选择每500次出一个损失,准确率
            print('[%d, %5d]: loss: %.3f , acc: %.2f %%'
                  % (epoch + 1, batch_idx + 1, running_loss/running_total, 100 * running_correct / running_total))
    return running_loss/running_total
    
```

```python
# L = criterion(torch.tensor([[0.1,2,3,4],[2,3,4,5]]),torch.tensor([1,2]))
```

```python
test_loader = DataLoader(val_dataset,10,shuffle=True)
```

```python
def test():
    correct = 0
    total = 0
    # 不计算梯度
    with torch.no_grad():
        for data in test_loader:
            inputs, labels = data['img'],data['target']
            prec = model(inputs)
            predicted = torch.argmax(torch.softmax(prec, dim=1),dim=1) 
            total += labels.size(0)
            # 张量之间的比较运算
            correct += (predicted == labels).sum().item()
    
    print('accuracy on test set: %d %% ' % (100 * correct / total))
    return correct/total
```

```python
loss = []
accuracy = []
for i in range(10):
    loss.append(train(i))
    accuracy.append(test())
```

```python
# Plot the loss curves for training
plt.plot(loss, color='b', label="training loss")
plt.xlabel("Number of Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
# Plot the accuracy curves for val
plt.plot(accuracy, color='b', label="accuracy")
plt.xlabel("Number of Epochs")
plt.ylabel("acc")
plt.legend()
plt.show()
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
#标准化图片尺寸
#大量不同大小和形状的图像。我们将定义一个调整大小的函数resize_ image，
from PIL import Image, ImageOps 

def resize_image(src_image, size=(128,128), bg_color="white"): 
    
    # 调整图像大小，使最长边与目标尺寸相匹配，短边按比例进行缩放
    src_image.thumbnail(size, Image.ANTIALIAS)
    
    # C创建一个背景
    new_image = Image.new("RGB", size, bg_color)
    
    # 将调整大小的图像粘贴到方形背景的中心
    new_image.paste(src_image, (int((size[0] - src_image.size[0]) / 2), int((size[1] - src_image.size[1]) / 2)))
  
    # 返回新图
    return new_image
```

```python
training_folder_name = './archive/data/natural_images'

# 存储处理后的图片
train_folder = './archive/working/data/natural_images'


# 为所有源图像创建调整大小的副本
size = (128,128)


if not os.path.exists(train_folder):
    os.makedirs(train_folder)

# 遍历文件夹下的每个子文件夹
print('转换图像...')
for root, folders, files in os.walk(training_folder_name):
    for sub_folder in folders:
        print('processing folder ' + sub_folder)
        # 创建子文件夹
        saveFolder = os.path.join(train_folder,sub_folder)
        if not os.path.exists(saveFolder):
            os.makedirs(saveFolder)
        # 遍历改子文件夹下的所有图片
        file_names = os.listdir(os.path.join(root,sub_folder))
        for file_name in file_names:
            #图片路径
            file_path = os.path.join(root,sub_folder, file_name)
            #打开图片
            image = Image.open(file_path)
            # 创建缩放版本并保存
            resized_image = resize_image(image, size)
            saveAs = os.path.join(saveFolder, file_name)
            resized_image.save(saveAs)

print('Done.')
```

```python
import torch
import torchvision
import torchvision.transforms as transforms
def load_dataset(data_path):

    # 随机修改图像，防止过拟合
    transformation = transforms.Compose([
        # 随机水平翻转
        transforms.RandomHorizontalFlip(0.5),
        # 随机垂直反转
        transforms.RandomVerticalFlip(0.3),
        # 转变为tensor类型，同时会把像素值归一化到【0,1】
        transforms.ToTensor(),
        # 标准化
        transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
    ])

    # 加载所有图像，并进行修改
    #将同一文件夹下的图片视为同一类，含有图片数据以及标签值
    full_dataset = torchvision.datasets.ImageFolder(
        root=data_path,
        transform=transformation
    )
    
    
    # 划分数据
    train_size = int(0.7 * len(full_dataset))
    test_size = len(full_dataset) - train_size
    train_dataset, test_dataset = torch.utils.data.random_split(full_dataset, [train_size, test_size])
    
    train_loader = torch.utils.data.DataLoader(
        train_dataset,
        batch_size=50,
        num_workers=0,
        shuffle=False
    )
    
    test_loader = torch.utils.data.DataLoader(
        test_dataset,
        batch_size=50,
        num_workers=0,
        shuffle=False
    )
        
    return train_loader, test_loader
```

```python
train_folder = './archive/working/data/natural_images'

train_loader, test_loader = load_dataset(train_folder)
batch_size = train_loader.batch_size
```

```python
#定义模型
class Net(nn.Module):
    
    
    def __init__(self, num_classes=3):
        super(Net, self).__init__()

        self.conv1 = nn.Conv2d(in_channels=3, out_channels=12, kernel_size=3, stride=1, padding=1)
        
        self.conv2 = nn.Conv2d(in_channels=12, out_channels=24, kernel_size=3, stride=1, padding=1)
        
        self.pool = nn.MaxPool2d(kernel_size=2)
        
        self.drop = nn.Dropout2d(p=0.2)

        self.fc = nn.Linear(in_features=32 * 32 * 24, out_features=num_classes)

    def forward(self, x):

        x = F.relu(self.pool(self.conv1(x))) 

        x = F.relu(self.pool(self.conv2(x)))  
        
        x = F.dropout(self.drop(x), training=self.training)
        
        # 展平
        x = x.view(-1, 32 * 32 * 24)
        # 传入全连接层做预测
        x = self.fc(x)
        # Return class probabilities via a log_softmax function 
        return x
    
device = "cpu"
if (torch.cuda.is_available()):
    # if GPU available, use cuda (on a cpu, training will take a considerable length of time!)
    device = "cuda"

# Create an instance of the model class and allocate it to the device
model = Net(num_classes=len(classes)).to(device)

print(model)
```

```python
def train(model, device, train_loader, optimizer, epoch):
    model.train()
    train_loss = 0
    print("Epoch:", epoch)
    for batch_idx, (data, target) in enumerate(train_loader):

        data, target = data.to(device), target.to(device)
        
        # Reset the optimizer
        optimizer.zero_grad()

        #计算模型输出
        output = model(data)
        
        # 损失
        loss = loss_criteria(output, target)

        # 记录损失
        train_loss += loss.item()
        
        # 反向传播
        loss.backward()
        optimizer.step()
        if batch_idx % 10==9:
            print('\tTraining batch {} Loss: {:.6f}'.format(batch_idx + 1, loss.item()))
            
    # 每个epoch的平均损失
    avg_loss = train_loss / (batch_idx+1)
    print('Training set: Average loss: {:.6f}'.format(avg_loss))
    return avg_loss
```

```python
def test(model, device, test_loader):
    #测试时不需要反向传播与drop
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        batch_count = 0
        for data, target in test_loader:
            batch_count += 1
            data, target = data.to(device), target.to(device)
            
            # 每个批次的输出
            output = model(data)
            
            # 每个批次的损失
            test_loss += loss_criteria(output, target).item()
            
            # 该批次的预测正确个数
            predicted = torch.argmax(torch.softmax(output,dim=1), 1)
            correct += torch.sum(target==predicted).item()

    # 该epoch的平均损失与正确率
    avg_loss = test_loss / batch_count
    print('Validation set: Average loss: {:.6f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        avg_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
    
    return avg_loss
```

```python
#指定优化器
optimizer = optim.Adam(model.parameters(), lr=0.01)

# 指定损失函数
loss_criteria = nn.CrossEntropyLoss()

epoch_nums = []
training_loss = []
validation_loss = []

# Train over 10 epochs (We restrict to 10 for time issues)
epochs = 10
print('Training on', device)
for epoch in range(1, epochs + 1):
        train_loss = train(model, device, train_loader, optimizer, epoch)
        test_loss = test(model, device, test_loader)
        epoch_nums.append(epoch)
        training_loss.append(train_loss)
        validation_loss.append(test_loss)
```

```python
plt.figure(figsize=(8,8))
plt.plot(epoch_nums, training_loss)
plt.plot(epoch_nums, validation_loss)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training', 'validation'], loc='upper right')
plt.show()
```

```python
#Defining Labels and Predictions
truelabels = []
predictions = []
model.eval()
print("Getting predictions from test set...")
for data, target in test_loader:
    data = data.to(device)
    for label in target.data.numpy():
        truelabels.append(label)
    output = torch.softmax(model(data),dim=1)
    for prediction in torch.argmax(output,dim=1):
        predictions.append(prediction.cpu().numpy()) 

# Plot the confusion matrix, 将每一样本预测为各个类的情况
cm = confusion_matrix(truelabels, predictions)
tick_marks = np.arange(len(classes))

df_cm = pd.DataFrame(cm, index = classes, columns = classes)
plt.figure(figsize = (7,7))
sns.heatmap(df_cm, annot=True, cmap=plt.cm.Blues, fmt='g')
plt.xlabel("Predicted Shape", fontsize = 20)
plt.ylabel("True Shape", fontsize = 20)
plt.show()
```
